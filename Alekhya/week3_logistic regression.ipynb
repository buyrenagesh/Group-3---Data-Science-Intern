### Data Cleaning 

### 1. Removing null values 

```python
df=df.dropna()
df.isnull().sum()
```

```python
df.shape
## Now there is no null value
```

### 1. Convert to lower case 

```python
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['text'].head()
```

### 2. Removing stopwards (unnecessary words which doesn't add any meaning to the text) 

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
df['text'] = df['text'].apply(lambda x : ' '.join([word for word in x.split() if not word in set(stopwords.words('english'))]))
```

### 3. Lemmatization & stemming (getting the root word)

```python
lemmatizer = WordNetLemmatizer()
df['text'] = df['text'].apply(lambda x : ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
ps = PorterStemmer()
df['text'] = df['text'].apply(lambda x : ' '.join([ps.stem(word) for word in x.split()]))
```

### 4. Removing Special Characters 

```python
def clean_tweets(tweet):
    # remove special characters 
    tweet = re.sub('[^ a-zA-Z0-9]', '', tweet)
    # remove Numbers
    tweet = re.sub('[0-9]', '', tweet)
    return tweet
df['text'] = df['text'].apply(clean_tweets)
```

```python
df['sentiment']=df['sentiment'].replace(['positive','negative','neutral'], [1,2,0]) 
```
